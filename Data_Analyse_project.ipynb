```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/Mall_Customers.csv")
print(df.head())
print(df.info())
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
print(df["Genre"])
```

```python
print(df["Age"])
df["Genter_category"] = df["Genre"].map({"Female":1,"Male":0})
```

```python
print("Average Age:",np.mean(df["Age"]))
```

```python
print("Average Annual Income:",np.mean(df["Annual Income (k$)"]))
```

```python
Genre_score=df.groupby("Genre")["Spending Score (1-100)"].mean()
print(Genre_score)
```

```python
# average income of each gender

Genre_income=df.groupby("Genre")["Annual Income (k$)"].mean()
print(Genre_income)
```

```python
sns.countplot(x="Genre", data=df)
plt.show()

sns.histplot(df["Age"], kde=True)
plt.show()
```

```python
sns.boxplot(x="Genre",y="Spending Score (1-100)",data=df)
plt.title("Genre vs spending Score")
plt.xlabel("Gender")
plt.ylabel(" Spending Score")
```
